## **clubList Function Definition:**
*   Returns a list of links to each team's squad page


```python
def clubList(url):
    response = requests.get(url + "/clubs" + yearCode)
    content = response.content
    soup = BeautifulSoup(content, "html.parser")
    cluburl = []
    clubs = []
    ul = soup.find_all(
        "ul",
        attrs={
            "class": "block-list-5 block-list-3-m block-list-1-s block-list-1-xs block-list-padding dataContainer"
        },
    )
    u = str(ul)
    soup2 = BeautifulSoup(u, "html.parser")
    for i, tags in enumerate(soup2.find_all("a")):
        cluburl.append(url + str(tags.get("href")))
    for i in range(0, len(cluburl)):
        cluburl[i] = cluburl[i].replace("overview", "squad")
    return cluburl
```

## **Function to scrape general data**

*   Access the dataframe using generalDf



```python
def generalAppend(url, playerov):
    playerOverview = playerov
    playerStat = []
    l = []
    for i in range(0, len(playerOverview)):
        response = requests.get(playerOverview[i])
        content = response.content
        soup = BeautifulSoup(content, "html.parser")
        playerNameTag = soup.find("div", attrs={"class": "name t-colour"})
        playerName = playerNameTag.text
        countryTag = soup.find("span", attrs={"class": "playerCountry"})
        country = countryTag.text
        jerseyNumberTag = soup.find("div", attrs={"class": "number t-colour"})
        jerseyNumber = jerseyNumberTag.text
        u = soup.find_all("div", attrs={"class": "playerOverviewAside u-show-mob"})
        u = str(u)
        soup2 = BeautifulSoup(u, "html.parser")
        t = soup2.find_all("div", attrs={"class": "info"})
        s = t[0]
        s = str(s)
        soup3 = BeautifulSoup(s, "html.parser")
        clubTag = soup3.find("a")
        club = clubTag.text
        club = str(club).strip()
        position = t[1].text
        l.append([club, playerName, jerseyNumber, country, position])
        # print(l[i])
    return l
```
